```python
def projection(sig, k, width, depth):
    if not isinstance(k, int):
        raise TypeError('k must be an integer')
    if k<0:
        raise NameError('projection level must be non-negative')
    if k==0:
        return np.array([1.])
    if k > depth:
        return np.zeros(width**k)
    if isinstance(sig, torch.DoubleTensor):
        sig = sig.numpy()
    k1 = sum([width**i for i in range(k)])
    k2 = sum([width**i for i in range(k+1)])
    return sig[k1:k2]
```

```python
def L2(tensor):
    return np.sqrt(sum([i**2 for i in tensor.flatten()]))
```

## Digit pathwise-prediction

```python
# load streams
with open('input.pickle', 'rb') as handle:
    X = pickle.load(handle, encoding='latin1')

# load corresponding digit (optional)
with open('output.pickle', 'rb') as handle:
    y = pickle.load(handle, encoding='latin1')
```

## Localised Regression

```python
stop_index = 5000
data = list(zip(X[:stop_index], y[:stop_index]))
random.shuffle(data)
X, y = zip(*data)
```

```python
# only retaining the first 80% of the information in the test set
percentage_to_retain = 0.8
X_input = [i[:int(percentage_to_retain*float(len(i))),:] for i in X]
X_missing_part = [i[int(percentage_to_retain*float(len(i))):,:] for i in X]
X_output = X
```

```python
split = int(.95*len(X))

X_input_train = X_input[:split]
X_output_train = X_output[:split]
y_train = y[:split]

X_input_test = X_input[split:]
X_output_test = X_output[split:]
y_test = y[split:]
```

```python
class LocalisedRegressionHomogeneous:
    
    def __init__(self, 
                 X_input_train, 
                 X_output_train,
                 y_train,
                 k_nn, 
                 depth,
                 width=2):  
        
        # Pre-compute signatures of training data
        self.X_input_train_sigs = [np.insert(iisignature.sig(path, depth), 0, 1.) for path in X_input_train]
        self.X_output_train_sigs = [np.insert(iisignature.sig(path, depth), 0, 1.) for path in X_output_train]

        self.X_input_train = X_input_train
        self.X_output_train = X_output_train
        self.y_train = y_train
        
        self.k_nn = k_nn
        self.width = width
        self.depth = depth
        
    def sig2sigtensor(self, sig):
        return np.insert(sig, 0, 2.)
        
    def norm(self, sig, width, depth):
        projection_norms = [(math.factorial(k) * L2(
                                                    projection(
                                                                te.tensor_log(self.sig2sigtensor(sig), depth)[1:], 
                                                                k, 
                                                                width, 
                                                                depth
                                                               )
                                                   )
                             )**(1./k)
                            for k in range(1, depth+1)
                           ]
        return (math.factorial(depth)**(-1./depth))*max(projection_norms)

    def hom_distance(self, stream_lhs, sig_rhs, width, depth):
        stream_lhs_inverse = np.flip(stream_lhs, axis=0)
        sig_tensor_lhs_inverse = self.sig2sigtensor(np.insert(iisignature.sig(stream_lhs_inverse, depth), 0, 1.))
        sig_tensor_rhs = self.sig2sigtensor(sig_rhs)    
        prod = te.tensor_multiply(sig_tensor_lhs_inverse, sig_tensor_rhs, depth)
        ans = self.norm(prod[1:], width, depth)
        if ans < 1e-07:
            return 0.
        return ans
        
    def train(self, x=None, y=None):
        pass
    
    def getKey(self, item):
        return item[0]
    
    def predict(self, x, y_x):
    
        # Multiprocessing
#         t = time.time()
#         distances = [self.hom_distance(x, siig, self.width, self.depth) for siig in self.X_input_train_sigs]
#         t4 = time.time()
#         print('time to compute distanes: {}'.format(t4-t))
        
        # Build list of useful items
        items = [(path_in, sig_in, path_out, sig_out, y) for path_in,sig_in,path_out,sig_out,y in zip(self.X_input_train, 
                                                                                                     self.X_input_train_sigs,
                                                                                                     self.X_output_train,
                                                                                                     self.X_output_train_sigs,
                                                                                                     self.y_train) if y==y_x]

        # compute signature of path x
        x_sig = np.insert(iisignature.sig(x, self.depth), 0, 1.)
        
        # pick k-NN
#         t = time.time()
#         final = sorted(items, key=self.getKey)
#         t7 = time.time()
#         print('time to sort: {}'.format(t7-t))
        
        final = items
        
        path_in_local = [i[0] for i in final[:self.k_nn]]
        sig_in_local = [i[1] for i in final[:self.k_nn]]
        path_out_local = [i[2] for i in final[:self.k_nn]]
        sig_out_local = [i[3] for i in final[:self.k_nn]]
        y_local = [i[4] for i in final[:self.k_nn]]
        
        # run linear regression locally on the selected k-NN
        lin_reg = LinearRegression()
        lin_reg.fit(sig_in_local, sig_out_local)
        
        return lin_reg.predict(np.array([x_sig])), path_in_local, path_out_local, y_local
```

```python
for depth in [2, 3, 4, 5, 6]:
    
    kk = tosig.sigdim(2, depth)
    how_many_samples = 4
    kks = range(kk+1, 4*kk, kk)

    for k in kks:
        model = LocalisedRegressionHomogeneous(X_input_train,
                                               X_output_train,
                                               y_train,
                                               k_nn=k, 
                                               depth=depth)

        lll = np.random.randint(0, len(y_test), how_many_samples)
#         lll = range(len(y_test))

        dims = (20, 15)
        fig, ax = plt.subplots(2, 2, figsize=dims)
        plt.suptitle('K-NN: {} --- depth: {}'.format(k, depth), fontsize=25)
        sss=0

        error = []

        for n in lll:

            ex = X_input_test[n]
            y_ex = y_test[n]

            pred, x_in_loc, x_out_loc, y_loc = model.predict(ex, y_ex)

            real_paths = X_output_test[n]
            increments = pred[0][1:3]
            starting_points = real_paths[0]
            predicted_paths = np.append(X_input_test[n], np.array([starting_points+increments]), axis=0)

            if sss == 0:
                a1, a2 = 0, 0
            if sss == 1:
                a1, a2 = 1, 0
            if sss == 2:
                a1, a2 = 0, 1
            if sss == 3:
                a1, a2 = 1, 1

            ax[a1, a2].plot(np.array([i[0] for i in predicted_paths]), np.array([i[1] for i in predicted_paths]), 'o--')
            ax[a1, a2].plot(np.array([i[0] for i in ex]), np.array([i[1] for i in ex]), marker='o')
            ax[a1, a2].plot(np.array([i[0] for i in real_paths]), np.array([i[1] for i in real_paths]))

            sss += 1 

        plt.show()
```

```python
for depth in [2, 3, 4, 5, 6]:
    
    kk = tosig.sigdim(2, depth)
    how_many_samples = 30
    squared_errors = []
    kks = range(kk+1, 4*kk)

    for k in kks:
        model = LocalisedRegressionHomogeneous(X_input_train,
                                               X_output_train,
                                               y_train,
                                               k_nn=k, 
                                               depth=depth)

        lll = np.random.randint(0, len(y_test), how_many_samples)
#         lll = range(len(y_test))

    #     dims = (20, 15)
    #     fig, ax = plt.subplots(2, 2, figsize=dims)
    #     plt.suptitle('K-NN: {}'.format(k), fontsize=25)
    #     sss=0

        error = []

        for n in lll:

            ex = X_input_test[n]
            y_ex = y_test[n]

            pred, x_in_loc, x_out_loc, y_loc = model.predict(ex, y_ex)

            real_paths = X_output_test[n]
            increments = pred[0][1:3]
            starting_points = real_paths[0]
            predicted_paths = np.append(X_input_test[n], np.array([starting_points+increments]), axis=0)

            error.append(np.sum((np.array([starting_points+increments]) - real_paths[-1])**2))

        squared_errors.append(np.mean(np.array(error)))

    #         if sss == 0:
    #             a1, a2 = 0, 0
    #         if sss == 1:
    #             a1, a2 = 1, 0
    #         if sss == 2:
    #             a1, a2 = 0, 1
    #         if sss == 3:
    #             a1, a2 = 1, 1

    #         ax[a1, a2].plot(np.array([i[0] for i in predicted_paths]), np.array([i[1] for i in predicted_paths]), marker='o')
    #         ax[a1, a2].plot(np.array([i[0] for i in ex]), np.array([i[1] for i in ex]), marker='o')
    #         ax[a1, a2].plot(np.array([i[0] for i in real_paths]), np.array([i[1] for i in real_paths]))

    #         sss += 1 

    #     plt.show()

    dims = (15, 8)
    fig, ax = plt.subplots(figsize=dims)
    plt.suptitle('Mean Squared Error for Localised Regression with depth {}'.format(depth), fontsize=20)
    plt.plot(kks, squared_errors, marker='o')
    plt.show()
```

```python
n_predictions = 5

for k in [7, 10, 15, 20, 50]:
    for d in [2, 3, 4, 5]:
        print('\n \n Predictions for k_nn = {}, depth = {}'.format(k, d))
        model = LocalisedRegressionHomogeneous(X_input_train, X_output_train, k_nn=k, depth=d)
        predicted_sigs = [model.predict(ex) for ex in X_input_test[:n_predictions]]
        real_paths = [p for p in X_output_test[:n_predictions]]
        increments = [p[0][1:3] for p in predicted_sigs]
        starting_points = [sp[0] for sp in real_paths]
        predicted_paths = [np.append(ex, np.array([sp+inc]), axis=0) for ex, sp, inc in zip(X_input_test[:n_predictions],
                                                                                            starting_points,
                                                                                            increments)]
        for item in range(n_predictions):
            plt.plot(np.array([i[0] for i in predicted_paths[item]]), np.array([i[1] for i in predicted_paths[item]]), marker='o')
            plt.plot(np.array([i[0] for i in real_paths[item]]), np.array([i[1] for i in real_paths[item]]))
            plt.title('prediction #{} with k_nn={}, depth={}'.format(item, k, d))
            plt.show()
```
