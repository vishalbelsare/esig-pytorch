```python
from sympy.ntheory import mobius
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
def bound_L2T(width, depth):
    a = (float(width)-1.)/float(width**(depth+1)-1)
    b = 0.
    for n in range(1, depth+1):
        for d in range(1, n+1):
            if n%d == 0:
                b += mobius(d)*(width**(n/d))
    return a*b
```

```python
h_map = {}
for width in range(2, 100):
    mom = {}
    for depth in range(2, 100):
        mom[depth] = bound_L2T(width, depth)
    h_map[width] = mom
```

```python
df = pd.DataFrame.from_dict(h_map)
```

```python
plt.figure(figsize=(12, 10))
seaborn.heatmap(df.values.astype('float'), cmap="YlGnBu", vmin=0.9999, vmax=1.0001)
plt.show()
```
